# Iris Classification
 A machine learning project aims to predict the species of iris flowers by utilizing measurements such as sepal width, sepal length, petal width, and petal length. The objective is to train a model on the Iris flower dataset, comprising 150 records with five attributes: Petal Length, Petal Width, Sepal Length, Sepal Width, and Class (Species) for identification. The dataset includes three types of Iris flower species. The samples for these three species are as follows:

# Model training and testing

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics

#Train - 70% of data
#Test - 30% of data
X = data_without_species  #input data
Y = data['species']       #output data
#splitting data for taining and testing
x_train, x_test,y_train, y_test = train_test_split(X,Y, test_size=0.3)
```

# Model 1 : Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_1 = LogisticRegression()
model_1.fit(x_train, y_train)
expected = Y
predicted = model_1.predict(X)
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("Logistic Regresssion Accuracy : ",model_1.score(x_test, y_test)*100,"%")
```

# Model 2 : k-nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
model_2 = KNeighborsClassifier()
model_2.fit(x_train, y_train)
predicted = model_2.predict(X)
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("KNeighbors Classifier Accuracy : ",model_2.score(x_test, y_test)*100,"%")
```

# Model 3 : Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
model_3 = DecisionTreeClassifier()
model_3.fit(x_train, y_train)
predicted = model_3.predict(X)
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
print("Decision Tree Classifier Accuracy : ",model_3.score(x_test, y_test)*100,"%")
```
